```python
#Question 7
#i select all the Lotus cars
df_lot = df[df.Make == 'Lotus']
```

```python
#ii select only columns 'Engine HP' and 'Engine Cylinders'
df_lot = df_lot[['Engine HP', 'Engine Cylinders']]
```

```python
#iii drop all duplicated rows
df_lot = df_lot.drop_duplicates()
df_lot
```

```python
#iv get underlying Numpy array
X = df_lot.values
X
```

```python
#v compute matrix-matrix mult
XTX = X.T.dot(X)
```

```python
#vi invert result
inv_XTX = np.linalg.inv(XTX)
```

```python
#vii create an array y
y =np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
```

```python
#viii multiply
w = inv_XTX.dot(X.T).dot(y)
```
